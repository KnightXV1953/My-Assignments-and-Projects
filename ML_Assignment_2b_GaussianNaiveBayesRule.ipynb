````python
# YOUR CODE HERE
#raise NotImplementedError()
import numpy as np
class GaussianBayes:
    ''' Implements the Gaussian Bayes For Classification... '''
    def __init__(self):
        self.mean = 0
        self.variance = 0
        self.priorprob = 0
        self.classes = 0
    def train(self, X, Y):

        self.classes = np.unique(Y)
        no_of_classes = len(self.classes)
        m = X.shape[1]

        self.mean = np.zeros((no_of_classes , m))
        self.variance = np.zeros((no_of_classes , m))
        self.priorprob = np.zeros(no_of_classes)


        for index , classes in enumerate(self.classes):
            data = X[Y==classes]
            self.mean[index , :] = np.mean(data, axis=0)
            self.variance[index , :] = np.var(data , axis = 0)
            self.priorprob[index] = data.shape[0] / X.shape[0]

        
        ''' Train the multiclass (or Binary) Bayes Rule using the given 
            X [m x n] data matrix and Y labels matrix'''
        
        # YOUR CODE HERE
        #raise NotImplementedError()
        
    def test(self, X):
        
        ''' Run the trained classifiers on the given set of examples 
            For each example, you should return probability and its assigned class
            Input: X of m x d
            Output:
            pclasses: predicted class of each example
            probability: probability of each example falling in that predicted class...
        '''
        
        nexamples, nfeatures=X.shape
        res=np.zeros((nexamples,len(self.classes)))
        pclasses=[0]*nexamples

        #print(res)

        for index , classes in enumerate(self.classes):
            mean = self.mean[index]
            variance = self.variance[index]  
            # Calculate the Gaussian probability for each feature
            prob = (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-0.5 * ((X - mean) ** 2) / variance)

            res[:, index] = np.prod(prob, axis=1) * self.priorprob[index]  # Multiply by prior probability for each class
    
        # Find the class with the maximum probability
        max_prob_indices = np.argmax(res, axis=1)
        pclasses = self.classes[max_prob_indices]      
        # YOUR CODE HERE
        #raise NotImplementedError()
        return np.array(pclasses), res
    
    def predict(self, X):
        return self.test(X)[0]    
        
````

# K-Fold Cross Validation

Since our dataset is not too big, to ensure that we are reporting true picture of our classifier, 
we will have to cross validate and report the mean accuracy across the folds to reflect the true 
picture of our classifier.

````python
# Now lets cross validate, and test the result...
nfolds=4
folds=t.generate_folds(X,Y,nfolds)

#now lets train and test on these folds...
totacc=[]
for k in range(nfolds):
    gb=GaussianBayes()
    gb.train(folds[k][0][:,2:],folds[k][1])
    pclasses,res=gb.test(folds[k][2][:,2:])
    acc=np.sum(pclasses==folds[k][3])/float(folds[k][3].shape[0])
    print ("[Info] Fold {} Accuracy = {}".format(k+1, acc))
    totacc.append(acc)

print (totacc, 'Mean Accuracy =', np.mean(totacc))
````
